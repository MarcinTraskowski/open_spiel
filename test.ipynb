```python
import pyspiel
import time

from open_spiel.python.algorithms import cfr
#from open_spiel.python import games
from open_spiel.python.algorithms import exploitability


game = pyspiel.load_game("python_patrol")
solver = cfr.CFRSolver(game)

start = time.time()
for i in range(100):
    solver.evaluate_and_update_policy()
end = time.time()
print(f"Elapsed: {end - start:.2f} seconds")


policy = solver.average_policy()

for state in policy.states:
    probs = policy.action_probabilities(state)
    print(state.information_state_string(state.current_player()), probs)


#conv = exploitability.exploitability(game, solver.average_policy())
```

```python
import pyspiel
import random

game = pyspiel.load_game("python_patrol")
state = game.new_initial_state()

random.seed(42)

print("Game:", game)
print("Initial state:", state)
print("-" * 40)

while not state.is_terminal():
    player = state.current_player()

    print("Player:", player)
    print("State:", state)

    if player == pyspiel.PlayerId.CHANCE:
        outcomes = state.chance_outcomes()
        print("Chance outcomes:", outcomes)

        actions, probs = zip(*outcomes)
        action = random.choices(actions, probs)[0]

    else:
        actions = state.legal_actions()
        print("Legal actions:", actions)

        action = random.choice(actions)

    print("Chosen action:", action)
    state.apply_action(action)
    print("-" * 40)

print("Terminal reached")
print("Returns:", state.returns())
```
