```python
import pyspiel
import random

game = pyspiel.load_game("patrol")
state = game.new_initial_state()

while not state.is_terminal():
    print(state)

    if state.current_player() >= 0:
        print("infostate:", state.information_state_string(state.current_player()))

    print("player:", state.current_player())

    actions = state.legal_actions()

    print("actions:", actions)
    
    action = random.choice(actions)

    print("chosen action:", action)
    state.apply_action(action)
    print()

print()
print("history:", state.history())
print("returns:", state.returns())
```

```python
import pyspiel
import time
from open_spiel.python.algorithms import cfr

game = pyspiel.load_game("patrol")  #  C++ game loaded in Python
solver = cfr.CFRSolver(game)

start = time.time()
for i in range(1000):
    solver.evaluate_and_update_policy()
end = time.time()
print(f"Time taken: {end - start}")

policy = solver.average_policy()

for state in policy.states:
    print(state, policy.action_probabilities(state))
```
